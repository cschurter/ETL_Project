```python
import pandas as pd
from sqlalchemy import create_engine

#from pymongo import MongoClient

#to insert data in MongoDB
#https://stackoverflow.com/questions/49221550/pandas-insert-a-dataframe-to-mongodb
```

```python
#Process File 1
events1_file = "Resources/events.csv"
events1_df = pd.read_csv(events1_file)
events1_df.head()
```

```python
#Drop Reviews Column
#Renamed column "Description" to "Name" 
events1_df = events1_df.drop(["Reviews"],axis=1).dropna()
events1_df = events1_df.rename(columns={'Description':'Name'})
events1_df.head()
```

```python
#Removing the first 3 words in Category column by using the split and join functions
events1_df['formatted_cat'] = events1_df.Category.map(lambda x: ' '.join(x.split(' ')[3:]))
events1_df.head()
```

```python
events1_df = events1_df.drop(["Category"],axis=1)
events1_df.head()
```

```python
#Clean File 1
events1_df = events1_df.rename(columns={'formatted_cat':'Category'})
events1_df.head()
```

```python
#Process File 2
events2_file = "Resources/TripAdvisor.csv"
events2_df = pd.read_csv(events2_file)
events2_df.head()
```

```python
#Clean File 2
events2_df = events2_df.drop(["# of Reviews"], axis=1)
events2_df.head()
```

```python
#Concatenate Files 1 and 2
```

```python
events1_events2_concat = pd.concat([events1_df,events2_df],ignore_index=True)
events1_events2_concat = events1_events2_concat.sort_values("Name")
events1_events2_concat.head()
```

```python
#Final Mission Bay events dataframe

mission_bay_events = events1_events2_concat.drop_duplicates(subset = ["Name","Address"])
mission_bay_events.head()
```

```python
#Process File 3 (Restaurants)
events3_file = "Resources/YelpExcel.xlsx"
events3_df = pd.read_excel(events3_file)
events3_df.head()
```

```python
events3_df = events3_df.drop(["Reviews"], axis=1)
events3_df.head()
```

```python
#Added "Category" column with "Food and Drink in San Diego" value
events3_df["Category"] = "Food & Drink in San Diego"
events3_df.head()
```

```python
#Final File 3
events3_df = events3_df.rename(columns = {"Categories": "Subcategory"})
events3_df = events3_df[["Name", "Address", "Phone", "Neighborhood", "Category", "Subcategory"]]
events3_df.head()
```
